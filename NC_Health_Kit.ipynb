```python
def glucosecheck():
  print("1. Fasting blood sugar level")
  print("2. one to two hours after meal")
  print("3. Hba1c")
  testtype =input("Choose a number from above 1 to 3: ")
  choosetest(testtype)
  
def diagnosis(unit,value,testtype):
  if unit == 'mg':
    #if the choice is fasting.
    if testtype == '1':
      if 70 <=value <= 99:
        print("Fasting blood sugare is within Normal range")
      
      #80–130 mg/dl (4.4–7.2 mmol/L)
      elif 80<=value<=130:
        print("Having diabetes")
      else:
        print("Invalid data")

    #if the choice is 1-2 hr after meals.
    elif testtype =='2':
      #Less than 140 mg/dl (7.8 mmol/L)
      if value<140:
        print("Normal")
      #Less than 180 mg/dl (10.0 mmol/L)
      elif 140<value<180:
        print("DM")
    
    #hba1c
    elif testtype =='3':
      print("test type 3")

    else :
      print('invalide response, try again')

  elif unit == 'mmol':

     if testtype == '1':
       if 3.9<=value <=5.5:
         print("Fasting blood sugare is within Normal range")
       elif 4.4<=value<=7.2:
         print("Having diabetes")
       else:
         print("Invalid data")
  
     #if the choice is 1-2 hr after meals.
     elif testtype =='2':
        if value <7.8:
          print("After meal test , Normal")
        elif 7.8<=value<10:
          print("DM")      
        else:
          print("Inavlide data")
  

  elif testtype == '3' and unit == "":
    if value < 5.7:
      print("HBa1c normal")
    elif 5.7 < value <7.0:
      print("HBA1c indicates DM")
    else:
      print("invalid response")
    #lessthan 5.7% Nl, ,7.0% DM


def choosetest(testtype):
    if testtype in ["1","2"]:
      unit = input("Choose the unit mg or mmol? Type the answer: ")
      value = float(input("What is the value? "))
      diagnosis(unit,value,testtype)
     
    elif testtype == "3":
      value = float(input("What is the value in percent"))
      diagnosis("",value,testtype)
      
    else:
      print("invalid response")



```

```python
def bpcheck():
  sbp =int(input("What is your systolic blood pressure: "))
  dbp =int(input("What is your diastolic blood pressure: "))
  

  if sbp <90 or dbp < 60:
    print ("Having low blood pressure")
  elif 90<= sbp <=120 or 60<=dbp<=80:
    print ("Ideal blood pressure")
  elif 120< sbp <=140 or 80 < dbp <=90:
    print ("Pre hypertension")
  elif 140<sbp< 190 or 90< dbp<100:
    print ("Hypertension")
  elif sbp>190 or dbp>100:
    print ("Hypertensive emergency")
  else:
    print("invalid result")


```
